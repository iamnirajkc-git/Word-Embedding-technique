```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
### Vocabulary size
voc_size=500
```

```python
#capture index of word
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding, pre add 0 at first where as in post 0 is added at last
## converting each vactor length to 8, will add 0 to make make vector length 8
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## 10 feature representation dimensions
dim=10
```

```python
## sent_length is 8 in our case
## Embedding is same like word2Vec
##our data is small so 10 dim is enough
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
#let see the feature representation of first vector in 10 dimensions
model.predict(embedded_docs[0])
```
